# Subplots
What if we instead wanted to plot sin(x) and cos(x) as separate plots, but side by side? This is where subplots(nrows, ncols) comes in handy. subplots(nrows, ncols) will return to us a figure object and a set of axes. To better understand subplots, think of a figure object as a large canvas where we can place a set of axes on. These set of axes are where we will plot our graphs on. Previously, we only had one single axis in our figure, and we were plotting 1 or many functions on that single axis. However, figures can have many axes, which we'll demonstrate using subplots(nrows, ncols). For example, in the code below we tell subplots(nrows, ncols) that we would like a figure object with 1 row and 2 columns (essentially an array with 2 elements, where each element is an axis that we can plot on). Since axes can be thought of as an array holding the figure axes, we can index it to specify what we'd like to plot in each one.

```python
# Example code from pdf

# one figure object with 1x2=2 axes
fig, axes = plt.subplots(nrows=1, ncols=2)

# plotting sin(x) in axes[0]. left image, or 1st element/axes in the axes array.
axes[0].plot(x, np.sin(x))
# setting the title of axes[0] to be y=sin(x)
axes[0].set_title('y=sin(x)')

# plotting cos(x) in axes[1]. right image, or 2nd element/axis in the axes array.
axes[1].plot(x, np.cos(x))
# setting the title of axes[1] to be y=cos(x)
axes[1].set_title('y=cos(x)')

plt.show()
```

```python
# My practical code

# one figure object with 2x1=2 axes
fig, axes = plt.subplots(nrows=2, ncols=1)

# plotting sin(x) in axes[0]. left image, or 1st element/axes in the axes array.
axes[0].plot(x, np.sin(x))
# setting the title of axes[0] to be y=sin(x)
axes[0].set_title('y=sin(x)')

# plotting cos(x) in axes[1]. right image, or 2nd element/axis in the axes array.
axes[1].plot(x, np.cos(x))
# setting the title of axes[1] to be y=cos(x)
axes[1].set_title('y=cos(x)')

plt.show()
```

```python
# Example code from pdf
x = np.linspace(1, 21)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))

axes[0][0].plot(x, x**2)
axes[0][1].plot(x, x)
axes[1][0].plot(x, np.log(x))
axes[1][1].plot(x, x**5)

plt.show()
```

# Custom Axes Range
Something else we'd sometimes like to do is change the x and y axis range. We can configure the ranges of the axes using the set_ylim and set_xlim, or axis('tight') for automatically getting "tightly fitted" axes ranges:

```python
# Example code from pdf
x = np.linspace(0, 5, 11)

fig, axes = plt.subplots(nrows=1, ncols=3)

# we can also plot multiple graphs using plt.plot(x1, y1, x2, y2).
axes[0].plot(x, x**2, x, x**4)
axes[0].set_title('default axis range')

axes[1].plot(x, x**2, x, x**4)
# tight happens to be the same here, but this will come in handy when you want to automatically align axes in
# fitted manner.
axes[1].axis('tight')
axes[1].set_title('tight axis range')

# notice the zoom in that happened.
axes[2].plot(x, x**2, x, x**4)
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[2].set_title("custom axes range");

plt.show()
```

```python
# My example code
# Adding 4th column
x = np.linspace(0, 5, 11)

fig, axes = plt.subplots(nrows=1, ncols=4)

# we can also plot multiple graphs using plt.plot(x1, y1, x2, y2).
axes[0].plot(x, x**2, x, x**4)
axes[0].set_title('default axis range')

axes[1].plot(x, x**2, x, x**4)
# tight happens to be the same here, but this will come in handy when you want to automatically align axes in
# fitted manner.
axes[1].axis('tight')
axes[1].set_title('tight axis range')

# notice the zoom in that happened.
axes[2].plot(x, x**2, x, x**4)
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[2].set_title("custom axes range");

# 4th column
axes[3].plot(x, x**2.5, x, x**4.5)
axes[3].set_ylim([0, 40])
axes[3].set_xlim([2, 3])
axes[3].set_title("custom axes range");
plt.show()
```
